```python
import pickle
import numpy as np
import pandas as pd
```

```python
EMBEDDINGS_PATH = '../../data/Full_emb.npy'
MAIN_DATA_PATH = '../../data/zappos_data/ut-zap50k-data/meta-data.csv'
DESCRIPTION_DATA_PATH = '../../data/descriptions_df_v0.csv'
```

```python
embeddings = np.load(EMBEDDINGS_PATH)
embeddings.shape
```

```python
embeddings = embeddings.astype(np.float32)
```

```python
desc_data_df = pd.read_csv(DESCRIPTION_DATA_PATH)
desc_data_df.head()
```

```python
main_data_df = pd.read_csv(MAIN_DATA_PATH)
main_data_df.head()
```

```python
data_df = desc_data_df[['CID', 'Url']]
data_df.head()
```

```python
cat_data_df = main_data_df[['CID', 'Category']]
cat_data_df.Category = cat_data_df.Category.astype('category')
cat_data_df.head()
```

```python
complete_data_df = pd.merge(data_df, cat_data_df, on='CID')
complete_data_df.head()
```

```python
complete_data_df['Class'] = complete_data_df.Category.cat.codes
complete_data_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df, train_embeddings, test_embeddings = train_test_split(complete_data_df, embeddings, test_size=0.25) 
```

```python
ZAPPOS_DIR = "../../data/zappos_char-cnn"
```

```python
import os
```

```python
TRAIN_DIR_PATH = os.path.join(ZAPPOS_DIR,'train')
TEST_DIR_PATH = os.path.join(ZAPPOS_DIR,'test')
```

```python
if not os.path.isdir(TRAIN_DIR_PATH):
    os.mkdir(TRAIN_DIR_PATH)
if not os.path.isdir(TEST_DIR_PATH):
    os.mkdir(TEST_DIR_PATH)
```

```python
train_embeddings.dtype
```

```python
with open(os.path.join(TRAIN_DIR_PATH, 'filenames.pickle'), 'wb') as f:
    pickle.dump(train_df.Url.tolist(), f)
    
with open(os.path.join(TRAIN_DIR_PATH, 'simple-embeddings.pickle'), 'wb') as f:
    pickle.dump(train_embeddings, f)

with open(os.path.join(TRAIN_DIR_PATH, 'class_info.pickle'), 'wb') as f:
    pickle.dump(train_df.Class.tolist(), f)
```

```python
with open(os.path.join(TEST_DIR_PATH, 'filenames.pickle'), 'wb') as f:
    pickle.dump(test_df.Url.tolist(), f)
    
with open(os.path.join(TEST_DIR_PATH, 'simple-embeddings.pickle'), 'wb') as f:
    pickle.dump(test_embeddings, f)
    
with open(os.path.join(TEST_DIR_PATH, 'class_info.pickle'), 'wb') as f:
    pickle.dump(test_df.Class.tolist(), f)
```
