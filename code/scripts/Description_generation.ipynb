```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from ast import literal_eval
import re
from tqdm import tqdm_notebook as tqdm
from random import shuffle
```

```python
data = pd.read_csv('../../data/shoes_v4.csv', index_col=0)
```

```python
data.head()
```

```python
for col in ['Closure', 'Gender', 'Material']:
    data[col] = data[col].apply(lambda x: literal_eval(x))
```

```python
data.ToeStyle = data.ToeStyle.apply(lambda x: str(x).split(';'))
data.Insole = data.Insole.apply(lambda x: str(x).split(';'))
```

```python
data.head()
```

```python
ToeStyle = data.apply(lambda x: pd.Series(x['ToeStyle']),axis=1).stack().reset_index(level=1, drop=True)
Insole = data.apply(lambda x: pd.Series(x['Insole']),axis=1).stack().reset_index(level=1, drop=True)
Closure = data.apply(lambda x: pd.Series(x['Closure']),axis=1).stack().reset_index(level=1, drop=True)
Gender = data.apply(lambda x: pd.Series(x['Gender']),axis=1).stack().reset_index(level=1, drop=True)
Material = data.apply(lambda x: pd.Series(x['Material']),axis=1).stack().reset_index(level=1, drop=True)
```

```python
insole_df = pd.DataFrame({'insole_s': Insole})
сlosure_df = pd.DataFrame({'closure_s': Closure})
gender_df = pd.DataFrame({'gender_s': Gender})
material_df = pd.DataFrame({'material_s': Material})
insole_df.head()
```

```python
stack_data = pd.concat((data,ToeStyle), axis=1, join='outer')
stack_data['ToeStyle'] = stack_data[0]
stack_data = stack_data.drop(columns = [0])
stack_data.head()
```

```python
#stack_data = pd.merge(stack_data, insole_df, how='right', left_on='lkey', right_on='rkey')
#stack_data['Insole'] = stack_data[0]
#stack_data = stack_data.drop(columns = [0])
stack_data = stack_data.join(insole_df)
stack_data['Insole'] = stack_data['insole_s']
stack_data = stack_data.drop(columns = ['insole_s'])
```

```python
stack_data = stack_data.join(сlosure_df)
stack_data['Closure'] = stack_data['closure_s']
stack_data = stack_data.drop(columns = ['closure_s'])
stack_data = stack_data.join(gender_df)
stack_data['Gender'] = stack_data['gender_s']
stack_data = stack_data.drop(columns = ['gender_s'])
stack_data = stack_data.join(material_df)
stack_data['Material'] = stack_data['material_s']
stack_data = stack_data.drop(columns = ['material_s'])
```

```python
stack_data[:10]
```

```python
stack_data['Color'] = stack_data['Color'].apply(lambda x: str(x).replace("['",""))
stack_data['Color'] = stack_data['Color'].apply(lambda x: str(x).replace("' '"," "))
stack_data['Color'] = stack_data['Color'].apply(lambda x: str(x).replace("']",""))
stack_data['Color']#re.sub("'[]", "", x))
```

```python
stack_data.head()
```

```python
descriptions_df = pd.DataFrame(columns=['CID', 'Url'])
```

```python
descriptions_df.head()
```

```python
stack_data = stack_data.fillna('')
```

```python
descriptions_df['CID'] = stack_data['CID']
descriptions_df['Url'] = stack_data['Url']
```

```python
stack_data.head()
```

```python
stack_data['Material'] = stack_data['Material'].apply(lambda x: str(x + ' material') if x!='' else '')
```

```python
stack_data['Gender'] = stack_data['Gender'].apply(lambda x: str('for '+x) if x!='' else '')
stack_data['HeelHeight'] = stack_data['HeelHeight'].apply(lambda x: str(x+' heel height') if x!='' else '')
stack_data['ToeStyle'] = stack_data['ToeStyle'].apply(lambda x: str(x+' style') if x!='' else '')
stack_data['Insole'] = stack_data['Insole'].apply(lambda x: str(x+' insole') if x!='' else '')
```

```python
stack_data['desc_list'] = stack_data[['Brand','SubCategory','Closure','Category','Material','Gender','HeelHeight', 'ToeStyle','Insole','Color']].values.tolist()
```

```python
stack_data.head()
```

```python
descriptions_df['splitted_desc'] = stack_data['desc_list'].reset_index()['desc_list']
```

```python
descriptions_df.head()
```

```python
NUM_DESC = 10
for i in range(NUM_DESC):
    descriptions_df.splitted_desc.apply(lambda x: shuffle(x))
    sent = descriptions_df.splitted_desc.values.flatten().tolist()
    descriptions_df['Descr'+str(i)] = [' '.join(t) for t in sent]
```

```python
descriptions_df.head()
```

```python
groupped_desc_list = descriptions_df.groupby(['CID','Url']).agg(list)
```

```python
groupped_desc_list = groupped_desc_list.reset_index()
```

```python
NUM_DESC = 10
for i in range(NUM_DESC):
    groupped_desc_list['Descr'+str(i)].apply(lambda x: shuffle(x))
    groupped_desc_list['Descr'+str(i)] = groupped_desc_list['Descr'+str(i)].apply(lambda x: x[0])
```

```python
groupped_desc_list.head()
```

```python
groupped_desc_list = groupped_desc_list.drop(columns=['splitted_desc'])
```
